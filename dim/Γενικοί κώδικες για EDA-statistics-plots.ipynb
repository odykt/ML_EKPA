```python
import pandas as pd          # Για την επεξεργασία δεδομένων (DataFrames)
import numpy as np           # Για τις αριθμητικές λειτουργίες (π.χ. υπολογισμοί IQR, ακραίες τιμές)
import seaborn as sns        # Για την οπτικοποίηση δεδομένων (π.χ. countplot, boxplot)
import matplotlib.pyplot as plt  # Για τη δημιουργία γραφημάτων (π.χ. για την εμφάνιση των plots)

# Φόρτωση του αρχείου overview.csv 
df_overview = pd.read_csv("overview.csv")

# Εμφάνιση των πρώτων γραμμών του dataframe
print(df_overview.head())

# Υπολογισμός βασικών στατιστικών για τις αριθμητικές στήλες
print(df_overview.describe())


# Ανάλυση κατανομής των κλάσεων για την στήλη 'Contrast'
print("Class Distribution for Contrast:")
print(df_overview['Contrast'].value_counts())  # Εκτύπωση των αποτελεσμάτων στο τερματικό
sns.countplot(data=df_overview, x='Contrast')
plt.title('Class Distribution for Contrast')
plt.xticks(fontsize=10)  
plt.show()

# Ανάλυση κατανομής των κλάσεων για την στήλη 'ContrastTag'
print("\nClass Distribution for ContrastTag:")
print(df_overview['ContrastTag'].value_counts())  # Εκτύπωση των αποτελεσμάτων στο τερματικό
sns.countplot(data=df_overview, x='ContrastTag')
plt.title('Class Distribution for ContrastTag')
plt.xticks(fontsize=8) 
plt.show()

# Δημιουργία boxplot για την ηλικία
sns.boxplot(x=df_overview['Age'])
plt.title('Boxplot για την Ηλικία')
plt.show()

# Υπολογισμός IQR για την Ηλικία
Q1 = df_overview['Age'].quantile(0.25)
Q3 = df_overview['Age'].quantile(0.75)
IQR = Q3 - Q1

# Προσδιορίζουμε τις ακραίες τιμές
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"Lower bound: {lower_bound}")
print(f"Upper bound: {upper_bound}")

# Εύρεση ακραίων τιμών
outliers = df_overview[(df_overview['Age'] < lower_bound) | (df_overview['Age'] > upper_bound)]
print(f"Ακραίες Τιμές: \n{outliers}")
```

```python
import pandas as pd          # Για την επεξεργασία δεδομένων (DataFrames)
import numpy as np           # Για τις αριθμητικές λειτουργίες (π.χ. NaN, υπολογισμοί)
import seaborn as sns        # Για την οπτικοποίηση δεδομένων (π.χ. heatmap)
import matplotlib.pyplot as plt  # Για την οπτικοποίηση δεδομένων (π.χ. scatter plot)
from sklearn.preprocessing import LabelEncoder  # Για την κωδικοποίηση ετικετών
from scipy import stats     # Για στατιστικές μεθόδους (π.χ. t-test για τον έλεγχο MCAR)


# Έλεγχος για ελλείπουσες τιμές σε κάθε στήλη
missing_values = df_overview.isnull().sum()
missing_percentage = (missing_values / len(df_overview)) * 100
missing_data = pd.DataFrame({'Missing Values': missing_values, 'Percentage': missing_percentage})
print(missing_data)

# Οπτικοποίηση ελλειπουσών τιμών με heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df_overview.isnull(), cbar=False, cmap='viridis')
plt.title('Heatmap Ελλειπουσών Τιμών')
plt.show()

# Εμφάνιση των γραμμών που περιέχουν missing values
missing_rows = df_overview[df_overview.isnull().any(axis=1)]
print("Γραμμές με Missing Values:")
print(missing_rows)

# Διερεύνηση των προτύπων ελλειπουσών τιμών - Ανάλυση Συσχέτισης (Correlation)

# Έλεγχος για MAR
# Δημιουργία στήλης για το αν η τιμή στο 'ContrastTag' είναι missing (True/False)
df_overview['ContrastTag_is_missing'] = df_overview['ContrastTag'].isnull()

# Μετατροπή των μη αριθμητικών στηλών σε αριθμητικές μέσω Label Encoding
encoder = LabelEncoder()


# Εφαρμογή LabelEncoder σε κάθε στήλη που περιέχει συμβολοσειρές
df_overview['raw_input_path_encoded'] = encoder.fit_transform(df_overview['raw_input_path'].astype(str))
df_overview['id_encoded'] = encoder.fit_transform(df_overview['id'].astype(str))
df_overview['tiff_name_encoded'] = encoder.fit_transform(df_overview['tiff_name'].astype(str))
df_overview['dicom_name_encoded'] = encoder.fit_transform(df_overview['dicom_name'].astype(str))

# Υπολογισμός συσχέτισης μεταξύ των στηλών
correlation = df_overview[['Age', 'Contrast', 'raw_input_path_encoded', 'id_encoded', 
                           'tiff_name_encoded', 'dicom_name_encoded', 'ContrastTag_is_missing']].corr()

# Εκτύπωση του πίνακα συσχέτισης στο terminal
print("Συσχέτιση (Correlation) για Age, Contrast, raw_input_path, id, tiff_name, dicom_name και Missing τιμές στο 'ContrastTag':")
print(correlation)



# Scatter plot για την συσχέτιση μεταξύ 'ContrastTag_is_missing' και 'Age'
plt.figure(figsize=(8,6))
plt.scatter(df_overview['Age'], df_overview['ContrastTag_is_missing'], alpha=0.5)
plt.title('Scatter Plot: Age vs Missing Values in ContrastTag')
plt.xlabel('Age')
plt.ylabel('ContrastTag is Missing (0 = No, 1 = Yes)')
plt.show()


# Έλεγχος για MNAR

# Έλεγχος για την ύπαρξη της στήλης 'ContrastTag_is_missing'
# Ελέγχουμε αν η στήλη 'ContrastTag_is_missing' υπάρχει στο DataFrame
print(df_overview.columns)

# Ελέγχουμε την κατανομή των missing τιμών στη στήλη 'ContrastTag'
print("Missing τιμές στη στήλη 'ContrastTag':")
print(df_overview['ContrastTag'].isnull().sum())

# Προβολή των γραμμών που έχουν missing τιμές στη στήλη 'ContrastTag'
missing_contrast_tag = df_overview[df_overview['ContrastTag'].isnull()]
print("\nΓραμμές με Missing Τιμές στη στήλη 'ContrastTag':")
print(missing_contrast_tag[['Unnamed: 0', 'Age', 'ContrastTag']])

# Δημιουργία της στήλης 'ContrastTag_is_missing'
# Δημιουργούμε τη στήλη που δηλώνει αν η τιμή είναι missing (True/False)
df_overview['ContrastTag_is_missing'] = df_overview['ContrastTag'].isnull()

# Δημιουργία πίνακα συχνοτήτων για τις missing τιμές του 'ContrastTag'
# Δημιουργία πίνακα συχνοτήτων για τις missing τιμές του ContrastTag σε σχέση με τις κατηγορίες του
missing_contrast_tag_crosstab = pd.crosstab(df_overview['ContrastTag_is_missing'], df_overview['ContrastTag'])

# Εκτύπωση του πίνακα συχνοτήτων για τις missing τιμές σε σχέση με τις κατηγορίες του 'ContrastTag'
print("\nΠίνακας Συχνότητας των Missing Τιμών για το 'ContrastTag':")
print(missing_contrast_tag_crosstab)

# Υπολογισμός ποσοστών για τις missing τιμές σε κάθε κατηγορία του 'ContrastTag'
missing_percentages = missing_contrast_tag_crosstab.div(missing_contrast_tag_crosstab.sum(axis=1), axis=0)

# Εκτύπωση των ποσοστών για τις missing τιμές σε κάθε κατηγορία του 'ContrastTag'
print("\nΠοσοστά Missing Τιμών για κάθε κατηγορία του 'ContrastTag':")
print(missing_percentages)


# Έλεγχος για MCAR
from fancyimpute import IterativeImputer
import missingno as msno  # Προσθήκη της βιβλιοθήκης missingno

# Δημιουργία ενός DataFrame με missing τιμές
df_overview = pd.DataFrame({
    'col1': [1, 2, np.nan, 4, 5],
    'col2': [5, np.nan, 7, 8, 9],
    'col3': [10, 11, 12, 13, np.nan]
})

# Εφαρμογή του IterativeImputer (μια μέθοδος imputation)
imputer = IterativeImputer()
df_imputed = imputer.fit_transform(df_overview)

# Επιστρέφει τα δεδομένα με τις missing τιμές να έχουν αντικατασταθεί
print(df_imputed)

# Οπτικοποίηση του μοτίβου των missing τιμών
msno.matrix(df_overview)  # Εδώ θα εμφανιστεί το μοτίβο των missing τιμών
plt.show()

# Επιστροφή ως pandas DataFrame
df_imputed_df = pd.DataFrame(df_imputed, columns=df_overview.columns)
print(df_imputed_df)

# MCAR test 
from scipy import stats

# Δημιουργία του DataFrame
df_overview = pd.DataFrame({
    'col1': [1, 2, np.nan, 4, 5],
    'col2': [5, np.nan, 7, 8, 9],
    'col3': [10, 11, 12, 13, np.nan]
})

# Για κάθε στήλη, ελέγχουμε αν τα missing είναι τυχαία
def mcar_test(df):
    significance_level = 0.05
    mcar = True  # Υποθέτουμε αρχικά ότι είναι MCAR

    for col in df.columns:
        print(f"\nΈλεγχος για στήλη: {col}")
        for other_col in df.columns:
            if col != other_col:
                missing = df[df[col].isnull()][other_col]
                not_missing = df[df[col].notnull()][other_col]

                # Αν έχουμε αρκετά δεδομένα και στις δύο ομάδες
                if len(missing.dropna()) > 0 and len(not_missing.dropna()) > 0:
                    t_stat, p_value = stats.ttest_ind(missing.dropna(), not_missing.dropna(), equal_var=False)
                    print(f"  Σύγκριση με {other_col}: p-value = {p_value:.4f}")
                    if p_value < significance_level:
                        mcar = False

    print("\nΤελικό συμπέρασμα:")
    if mcar:
        print("Τα missing values είναι πιθανό να είναι MCAR (Missing Completely At Random).")
    else:
        print("Τα missing values ΔΕΝ είναι MCAR.")
        
# Τρέχουμε το τεστ
mcar_test(df_overview)
```

```python
# Στατιστική ανάλυση .npz αρχείου
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from skimage.transform import resize

# Φόρτωση του .npz αρχείου με allow_pickle=True
data = np.load('full_archive.npz', allow_pickle=True)

# Εμφάνιση των πρώτων στοιχείων του 'image' και 'idx'
print("First few elements of 'image':", data['image'][:5])
print("First few elements of 'idx':", data['idx'][:5])

# Αποθήκευση των εικόνων και των ετικετών
images = data['image']  # Οι εικόνες
targets = data['idx']   # Οι ετικετές (ή οι κλάσεις)

# Ελέγχουμε τις διαστάσεις των δεδομένων
print(f"Shape of images: {images.shape}")
print(f"Shape of targets: {targets.shape}")

# Ελέγχουμε τις διαστάσεις κάθε εικόνας χειροκίνητα
image_shapes = [image.shape for image in images]

# Εκτύπωση των πρώτων 5 σχημάτων εικόνας για έλεγχο
print("First few image shapes:", image_shapes[:5])

# Αν υπάρχουν διαφορετικά μεγέθη, εντοπίζουμε τα μοναδικά σχήματα
unique_shapes = list(set(image_shapes))
print(f"Unique shapes of images: {unique_shapes}")

# Ελέγχουμε αν όλες οι εικόνες έχουν το ίδιο μέγεθος
if len(unique_shapes) > 1:
    print("Προειδοποίηση: Υπάρχουν εικόνες με διαφορετικά μεγέθη. Ελέγξτε τις παρακάτω μοναδικές διαστάσεις.")
else:
    print("Όλες οι εικόνες έχουν το ίδιο μέγεθος.")

# Αν οι εικόνες δεν είναι όλες 512x512, μπορούμε να τις προσαρμόσουμε ή να διαγράψουμε τις μη έγκυρες
# Εδώ επιλέγουμε να διατηρήσουμε μόνο τις εικόνες με μέγεθος (512, 512)
images = [image for image in images if image.shape == (512, 512)]

# Επανακαθορισμός του πίνακα εικόνας
print(f"New number of images: {len(images)}")

# Επανασχηματισμός όλων των εικόνων σε μέγεθος (512, 512)
resized_images = [resize(image, (512, 512), mode='reflect') for image in images]

# Ελέγχουμε τις διαστάσεις των πρώτων εικόνων για επιβεβαίωση
print("Resized images shapes:", [image.shape for image in resized_images][:5])

# Κάνουμε flatten τις εικόνες σε 1D για κάθε εικόνα
flattened_images = np.array([image.flatten() for image in resized_images])

# Δημιουργούμε το DataFrame για να υπολογίσουμε τα στατιστικά
df_images = pd.DataFrame(flattened_images)

# Υπολογισμός βασικών στατιστικών μέτρων για τα pixels σε όλες τις εικόνες
statistics = df_images.describe()

# Ελέγχουμε την δομή του στατιστικού πίνακα
print("\nStatistics (full):")
print(statistics)

# Υπολογισμός μέσης τιμής και τυπικής απόκλισης
mean_values = df_images.mean()  # Μέση τιμή για κάθε pixel
std_values = df_images.std()    # Τυπική απόκλιση για κάθε pixel

print("\nΜέση τιμή των pixels:")
print(mean_values)

print("\nΤυπική απόκλιση των pixels:")
print(std_values)

# Ελάχιστη και μέγιστη τιμή
min_values = df_images.min()
max_values = df_images.max()
print("\nΕλάχιστη τιμή των pixels:", min_values)
print("Μέγιστη τιμή των pixels:", max_values)

# 25%, 50%, 75% Percentiles
percentiles_25 = statistics.loc['25%']
percentiles_50 = statistics.loc['50%']  # Διάμεσος
percentiles_75 = statistics.loc['75%']
print("\n25th Percentile: ", percentiles_25)
print("50th Percentile (Median): ", percentiles_50)
print("75th Percentile: ", percentiles_75)

# Ανάλυση της κατανομής των κλάσεων-στόχων (class distribution)
class_distribution = pd.Series(targets).value_counts()

# Φιλτράρισμα των κλάσεων που έχουν περισσότερα από 1 δείγμα
filtered_class_distribution = class_distribution[class_distribution > 1]

# Έλεγχος αν υπάρχουν κλάσεις με περισσότερα από 1 δείγμα
if filtered_class_distribution.empty:
    print("Δεν υπάρχουν κλάσεις με περισσότερα από 1 δείγμα.")
else:
    # Δημιουργία του bar plot για τις πιο συχνές κλάσεις
    top_classes = filtered_class_distribution.head(10)
    plt.figure(figsize=(10, 6))
    top_classes.plot(kind='bar', color='purple')
    plt.title('Top 10 Most Frequent Classes (with more than 1 sample)')
    plt.xlabel('Class')
    plt.ylabel('Count')
    plt.show()

# Ομαδοποίηση των κλάσεων με λιγότερα από 2 δείγματα σε μία κατηγορία "Άλλες"
class_distribution_grouped = class_distribution[class_distribution > 1]

# Ομαδοποίηση των κλάσεων με 1 δείγμα ή λιγότερο στην κατηγορία "Other"
other_classes = class_distribution[class_distribution <= 1].sum()
class_distribution_grouped = pd.concat([class_distribution_grouped, pd.Series([other_classes], index=['Other'])])

# Δημιουργία του bar plot για την ομαδοποιημένη κατανομή κλάσεων
plt.figure(figsize=(10, 6))
class_distribution_grouped.plot(kind='bar', color='green')
plt.title('Grouped Class Distribution (with "Other" Category)')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()

# Δειγματοληψία 50% των δεδομένων για πιο γρήγορη ανάλυση
sample_class_distribution = class_distribution.sample(frac=0.5, random_state=42)

# Δημιουργία του bar plot για τη δειγματοληψία
plt.figure(figsize=(10, 6))
sample_class_distribution.plot(kind='bar', color='orange')
plt.title('Sampled Class Distribution (50%)')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()

# Διεξαγωγή ανάλυσης για ακραίες τιμές (Outliers)
Q1 = df_images.quantile(0.25)
Q3 = df_images.quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Εντοπίζουμε τις ακραίες τιμές
outliers = ((df_images < lower_bound) | (df_images > upper_bound)).sum()

# Εκτύπωση των ακραίων τιμών
print("\nΑκραίες τιμές (Outliers) σε κάθε στήλη:")
print(outliers)
```
